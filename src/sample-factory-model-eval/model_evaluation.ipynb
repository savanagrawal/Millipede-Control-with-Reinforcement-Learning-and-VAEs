# Comparison of APPO Variations in Sample Factory

We are importing necessary Python libraries for our analysis. This includes:

- `pandas` for data manipulation,
- `matplotlib` and `numpy` for data visualization and mathematical operations,
- `wget` for downloading files from the internet, 
- and `matplotlib.style` and `matplotlib` for setting the style of our plots.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.style
import matplotlib as mpl
import wget
```

## Reading Trained Data into DataFrames

In this cell, we're loading multiple .csv files into pandas DataFrames. The .csv files contain data about different runs of an experiment involving the movement of an ant in the x-direction. Each DataFrame holds specific data about these runs, such as the reward, kl divergence, loss, and the fraction of clipped steps.


```python
df1 = pd.read_csv('./assets/run-ant_x_direction_pbt_task_.summary_0-tag-reward_reward.csv')
df2 = pd.read_csv('./assets/run-ant_x_direction_rnn_task_.summary_0-tag-reward_reward.csv')
df3 = pd.read_csv('./assets/run-ant_x_direction_sync_task_.summary_0-tag-reward_reward.csv')
df4 = pd.read_csv('./assets/run-ant_x_direction_task_.summary_0-tag-reward_reward.csv')

df5 = pd.read_csv('./assets/run-ant_x_direction_pbt_task_.summary_0-tag-train_kl_divergence.csv')
df6 = pd.read_csv('./assets/run-ant_x_direction_rnn_task_.summary_0-tag-train_kl_divergence.csv')
df7 = pd.read_csv('./assets/run-ant_x_direction_sync_task_.summary_0-tag-train_kl_divergence.csv')
df8 = pd.read_csv('./assets/run-ant_x_direction_task_.summary_0-tag-train_kl_divergence.csv')

df9 = pd.read_csv('./assets/run-ant_x_direction_pbt_task_.summary_0-tag-train_loss.csv')
df10 = pd.read_csv('./assets/run-ant_x_direction_rnn_task_.summary_0-tag-train_loss.csv')
df11 = pd.read_csv('./assets/run-ant_x_direction_sync_task_.summary_0-tag-train_loss.csv')
df12 = pd.read_csv('./assets/run-ant_x_direction_task_.summary_0-tag-train_loss.csv')

df13 = pd.read_csv('./assets/run-ant_x_direction_pbt_task_.summary_0-tag-train_fraction_clipped.csv')
df14 = pd.read_csv('./assets/run-ant_x_direction_rnn_task_.summary_0-tag-train_fraction_clipped.csv')
df15 = pd.read_csv('./assets/run-ant_x_direction_sync_task_.summary_0-tag-train_fraction_clipped.csv')
df16 = pd.read_csv('./assets/run-ant_x_direction_task_.summary_0-tag-train_fraction_clipped.csv')

```

## Steps vs Rewards Plot

We are plotting the reward received at different steps of the run for various configurations (PBT, RNN, Synchronous, Default Parameters). This gives us a comparative look at the performance of these configurations.

```python
plt.figure(figsize=(15, 6))
plt.plot(df1['Step'], df1['Value'], label='PBT True')
plt.plot(df2['Step'], df2['Value'], label='RNN True')
plt.plot(df3['Step'], df3['Value'], label='Synchronous True')
plt.plot(df4['Step'], df4['Value'], label='Deafult Parameters')


# Adding labels, title, and grid
plt.xlabel('Steps')
plt.ylabel('Reward')
plt.title('Steps vs Reward')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## Steps vs KL Divergence Plot

Here we are plotting the Kullback-Leibler (KL) Divergence at different steps of the run for various configurations (PBT, RNN, Synchronous, Default Parameters). The KL Divergence measures how one probability distribution diverges from a second, expected probability distribution. 

```python
plt.figure(figsize=(15, 6))
plt.plot(df5['Step'], df5['Value'], label='PBT True')
plt.plot(df6['Step'], df6['Value'], label='RNN True')
plt.plot(df7['Step'], df7['Value'], label='Synchronous True')
plt.plot(df8['Step'], df8['Value'], label='Deafult Parameters')


# Adding labels, title, and grid
plt.xlabel('Steps')
plt.ylabel('KL Divergence')
plt.title('Steps vs KL Divergence')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## Steps vs Loss Plot

This cell contains code to plot the Loss incurred at different steps of the run for various configurations (PBT, RNN, Synchronous, Default Parameters). The loss is a measure of the difference between the model's predictions and the actual data.


```python
plt.figure(figsize=(15, 6))
plt.plot(df9['Step'], df9['Value'], label='PBT True')
plt.plot(df10['Step'], df10['Value'], label='RNN True')
plt.plot(df11['Step'], df11['Value'], label='Synchronous True')
plt.plot(df12['Step'], df12['Value'], label='Deafult Parameters')


# Adding labels, title, and grid
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Steps vs Loss')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## Steps vs Fraction Clipped Plot

In this cell, we're plotting the 'Fraction Clipped' at different steps of the run for various configurations (PBT, RNN, Synchronous, Default Parameters). Fraction clipped refers to the fraction of steps where the policy update was clipped due to large deviation from the current policy during optimization. 


```python
plt.figure(figsize=(15, 6))
plt.plot(df13['Step'], df13['Value'], label='PBT True')
plt.plot(df14['Step'], df14['Value'], label='RNN True')
plt.plot(df15['Step'], df15['Value'], label='Synchronous True')
plt.plot(df16['Step'], df16['Value'], label='Deafult Parameters')


# Adding labels, title, and grid
plt.xlabel('Steps')
plt.ylabel('Fraction Clipped')
plt.title('Steps vs Fraction Clipped')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## Final Analysis

We define a function named `create_subplots()`. This function generates subplots comparing reward, KL Divergence, loss, and fraction clipped for all configurations (PBT, RNN, Synchronous, Default Parameters) in a 2x2 grid so we can analyse them together.


```python
def create_subplots():
    fig, ax = plt.subplots(2, 2, figsize=(13, 8))

    ax[0, 0].plot(df1['Step'], df1['Value'], label='PBT True')
    ax[0, 0].plot(df2['Step'], df2['Value'], label='RNN True')
    ax[0, 0].plot(df3['Step'], df3['Value'], label='Synchronous True')
    ax[0, 0].plot(df4['Step'], df4['Value'], label='Default Parameters')
    ax[0, 0].set_title('Steps vs Reward')
    ax[0, 0].set_xlabel('Steps')
    ax[0, 0].set_ylabel('Reward')

    ax[0, 1].plot(df5['Step'], df5['Value'], label='PBT True')
    ax[0, 1].plot(df6['Step'], df6['Value'], label='RNN True')
    ax[0, 1].plot(df7['Step'], df7['Value'], label='Synchronous True')
    ax[0, 1].plot(df8['Step'], df8['Value'], label='Default Parameters')
    ax[0, 1].set_title('Steps vs KL Divergence')
    ax[0, 1].set_xlabel('Steps')
    ax[0, 1].set_ylabel('KL Divergence')

    ax[1, 0].plot(df9['Step'], df9['Value'], label='PBT True')
    ax[1, 0].plot(df10['Step'], df10['Value'], label='RNN True')
    ax[1, 0].plot(df11['Step'], df11['Value'], label='Synchronous True')
    ax[1, 0].plot(df12['Step'], df12['Value'], label='Default Parameters')
    ax[1, 0].set_title('Steps vs Loss')
    ax[1, 0].set_xlabel('Steps')
    ax[1, 0].set_ylabel('Loss')

    ax[1, 1].plot(df13['Step'], df13['Value'], label='PBT True')
    ax[1, 1].plot(df14['Step'], df14['Value'], label='RNN True')
    ax[1, 1].plot(df15['Step'], df15['Value'], label='Synchronous True')
    ax[1, 1].plot(df16['Step'], df16['Value'], label='Default Parameters')
    ax[1, 1].set_title('Steps vs Fraction Clipped')
    ax[1, 1].set_xlabel('Steps')
    ax[1, 1].set_ylabel('Fraction Clipped')

    # Add main title
    plt.suptitle('Comparative Performance of APPO Variations in Sample Factory', fontsize=16)

    # Add a legend for the whole figure
    handles, labels = ax[0,0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper right')

    plt.tight_layout()  # Adjust subplot parameters to give specified padding
    plt.savefig('AppoVariations.png', dpi=300)
    plt.show()

create_subplots()
```
